```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_2020_2021_df =pd.concat([sales_2020_df,sales_2021_df],axis = 0)
sales_2020_2021_df = sales_2020_2021_df.reset_index(drop = True)

sales_2020_2021_df.tail(10)
```

```python
# Check if any values are null.
sales_2020_2021_df.isnull().sum()
```

```python
# Check the data type of each column
sales_2020_2021_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_2020_2021_df["invoice_date"]= pd.to_datetime(sales_2020_2021_df['invoice_date'], format = '%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_2020_2021_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_df = sales_2020_2021_df.groupby(['region','state','city'])['units_sold'].sum()
region_df = region_df.reset_index()

region_df = region_df.rename(columns={'units_sold':'Total_Products_Sold'})
# Show the top 5 results.
region_df = region_df.sort_values('Total_Products_Sold', ascending= False)
region_df.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
region_pivot_df = pd.pivot_table(sales_2020_2021_df,
                                 index=['region','state','city'],
                                 values='units_sold',
                                 aggfunc='sum')
region_pivot_df= region_pivot_df.reset_index()


# Rename the "units_sold" column to "Total_Products_Sold"
region_pivot_df = region_pivot_df.rename(columns={'units_sold':'Total_Products_Sold'})
region_pivot_df = region_pivot_df.sort_values("Total_Products_Sold", ascending= False)
# Show the top 5 results.
region_pivot_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_df = sales_2020_2021_df.groupby(['region','state','city'])['total_sales'].sum()
region_df = region_df.reset_index()
region_df = region_df.rename(columns={'total_sales':"Total_Sales"})

# Show the top 5 results.
region_df = region_df.sort_values('Total_Sales',ascending= False)
region_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
region_pivot_df = pd.pivot_table(sales_2020_2021_df,
                                 index=['region','state','city'],
                                 values='total_sales',
                                 aggfunc='sum'
                                 )
region_pivot_df =region_pivot_df.reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
region_pivot_df = region_pivot_df.rename(columns={'total_sales':'Total_Sales'})

# Show the top 5 results.
region_pivot_df = region_pivot_df.sort_values('Total_Sales', ascending= False)
region_pivot_df.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_df = sales_2020_2021_df.groupby(['retailer','region','state','city'])['total_sales'].sum()
retailer_df = retailer_df.reset_index()

retailer_df = retailer_df.rename(columns={'total_sales':'Total_Sales'})
# Show the top 5 results.
retailer_df = retailer_df.sort_values('Total_Sales', ascending= False)
retailer_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_pivot_df = pd.pivot_table(sales_2020_2021_df,
                                   index=['retailer','region','state','city'],
                                   values='total_sales',
                                   aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
retailer_pivot_df = retailer_pivot_df.rename(columns={'total_sales':'Total_Sales'})

# Show the top 5 results.
retailer_pivot_df = retailer_pivot_df.sort_values('Total_Sales',ascending=False)
retailer_pivot_df.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_fw_sales_df = sales_2020_2021_df[sales_2020_2021_df['product'] == "Women's Athletic Footwear"]
women_fw_sales_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_retailer_groupby = women_fw_sales_df.groupby(['retailer','region','state','city'])['units_sold'].sum()
women_retailer_groupby = women_retailer_groupby.reset_index()
women_retailer_groupby = women_retailer_groupby.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
women_retailer_groupby = women_retailer_groupby.sort_values('Womens_Footwear_Units_Sold',ascending= False)
women_retailer_groupby.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_retailer_pivot_table = pd.pivot_table(women_fw_sales_df,
                                            index=['retailer','region','state','city'],
                                            values='units_sold',
                                            aggfunc='sum')
women_retailer_pivot_table = women_retailer_pivot_table.reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_retailer_pivot_table = women_retailer_pivot_table.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
women_retailer_pivot_table = women_retailer_pivot_table.sort_values('Womens_Footwear_Units_Sold',ascending= False)
women_retailer_pivot_table.head()
```
